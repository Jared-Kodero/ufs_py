```python
with open("modules.txt", "r") as f:
    lines = f.readlines()
    stop_idx = None
    for i, line in enumerate(lines):
        if "Where" in line:
            stop_idx = i

    lines = lines[:stop_idx]

    new_lines = []
    for line in lines:
        if line.startswith("-------"):
            continue
        # if line.strip().startswith(""):
        #     continue
        if "(D)" in line:
            new_line = line.replace("(D)", "").strip().split()
            new_lines.append(",".join(new_line))
        else:
            new_line = line.strip().split()
            if not new_line:
                continue
            new_lines.append(",".join(new_line))


new_lines.sort()
modules = {}

for entry in new_lines:
    # split on comma in case multiple modules per string
    parts = [p.strip() for p in entry.split(",") if p.strip()]

    for mod in parts:
        if "/" not in mod:
            continue  # skip malformed entries

        name, version = mod.split("/", 1)

        if name not in modules:
            modules[name] = set()

        modules[name].add(version)


modules = {k: sorted(v) for k, v in modules.items()}
import pprint

pprint.pprint(modules)
```
